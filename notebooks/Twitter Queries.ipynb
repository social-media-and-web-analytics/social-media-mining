# Tweet data
Solutions to the exercises below can be found by reading through the documentation on the [Tweet object model](https://developer.twitter.com/en/docs/twitter-api/data-dictionary/object-model/tweet).

## Exercise 08
Run the code below to search for recent Tweets containing the word "twitter" and to ensure Tweets are returned containing their `id`, `text`, and `created_at` fields.


```python
query = 'twitter'
fields = [
    'id',
    'text',
    'created_at',
]
tweets = client.search_recent_tweets(query=query, max_results=10,tweet_fields=fields)
for tweet in tweets.data:
    print(f'Tweet ID: {tweet.id}')
    print(f'Created: {tweet.created_at}\n')
    print(tweet.text)
    print('\n')
```

## Exercise 09
Update the `fields` variable in the code below to search for recent tweets that contain the word "twitter" and to ensure Tweets are returned containing their `id`, `text`, `created_at`, and `public_metrics` fields.

```python
query = 'twitter'
#fields = [] # (SOLUTION)
fields = [
    'id',
    'text',
    'created_at',
    'public_metrics',
]

tweets = client.search_recent_tweets(query=query, max_results=10,tweet_fields=fields)
for tweet in tweets.data:
    print(f'Tweet ID: {tweet.id}')
    print(f'Created: {tweet.created_at}\n')
    print(tweet.text)
    print(tweet.public_metrics)
    print('\n')
```
